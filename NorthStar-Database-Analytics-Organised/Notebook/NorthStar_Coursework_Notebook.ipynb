## 1. Import Libraries

```
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3
print('Ready')
```

## 2. Load Sample Data

```
df = pd.read_csv('../Data/sample_northstar_data.csv')
df['profit'] = df['revenue'] - df['cost']
display(df)
```

## 3. Python Analytics

```
zone_summary = df.groupby('city_zone').agg(total_journeys=('journey_id','count'), avg_delay=('delay_minutes','mean'), total_profit=('profit','sum')).reset_index()
display(zone_summary)
```

## 4. Visualisation

```
plt.figure(figsize=(8,5))
plt.bar(zone_summary['city_zone'], zone_summary['avg_delay'])
plt.xlabel('City Zone')
plt.ylabel('Average Delay Minutes')
plt.title('Average Delay by City Zone')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 5. SQL Analytics

```
conn = sqlite3.connect(':memory:')
df.to_sql('journeys', conn, index=False, if_exists='replace')
query = '''SELECT city_zone, COUNT(*) AS total_journeys, ROUND(AVG(delay_minutes),2) AS avg_delay FROM journeys GROUP BY city_zone ORDER BY avg_delay DESC;'''
display(pd.read_sql_query(query, conn))
```

## 6. SQL Optimisation

```
cursor = conn.cursor()
cursor.execute('CREATE INDEX idx_zone ON journeys(city_zone);')
cursor.execute('CREATE INDEX idx_status ON journeys(delivery_status);')
explain = '''EXPLAIN QUERY PLAN SELECT city_zone, COUNT(*) FROM journeys WHERE delivery_status='Delayed' GROUP BY city_zone;'''
display(pd.read_sql_query(explain, conn))
```
